```python
# Algorithem d'insertion avec calcul de complexité
def tri_insertion(tab):
    # Compteurs pour la complexité
    comparaisons = 0
    echanges = 0

    for i in range(1, len(tab)):
        y = tab[i]
        j = i - 1
        
        
        while j >= 0:
            comparaisons += 1  # Incrémenter à chaque comparaison
            if y < tab[j]:
                tab[j + 1] = tab[j]  
                echanges += 1  # Compter comme un échange
            else:
                break  
            j -= 1
        
        tab[j + 1] = y  

    print(f"Nombre total de comparaisons: {comparaisons}")
    print(f"Nombre total d'échanges: {echanges}")
    return tab

# Test du code avec un tableau donné
tab = [6, 2, 7, 1, 9, 5, 10,100,30,23,116]

tri_insertion(tab)

print("Le tableau trié est :")
for i in range(len(tab)):
    print("%d" % tab[i])

```

```python
import time
import random

n_values = [1000, 2000, 4000, 8000, 16000, 32000]
times = []

# Mesurer le temps d'exécution pour chaque taille de tableau
for n in n_values:
    
    tab = list(range(n, 0, -1))
    
    
    start_time = time.time()
    tri_insertion(tab)
    end_time = time.time()
    
    
    execution_time = end_time - start_time
    times.append(execution_time)

    print(f"Taille du tableau : {n}, Temps d'exécution : {execution_time:.5f} secondes")
```

```python
import matplotlib.pyplot as plt


plt.plot(n_values, times, marker='o', linestyle='-', color='b', label="Temps d'exécution")
plt.xlabel("Taille du tableau (n)")
plt.ylabel("Temps d'exécution (secondes)")
plt.title("Temps d'exécution T(n) en fonction de la taille du tableau")
plt.legend()
plt.grid(True)
plt.show()
```

```python
def tri_selection(tab):
    # Compteurs pour la complexité
    comparaisons = 0
    echanges = 0

    for i in range(len(tab)):
        min = i

        
        for j in range(i + 1, len(tab)):
            comparaisons += 1  
            if tab[j] < tab[min]:
                min = j

        
        if min != i:
            tab[i], tab[min] = tab[min], tab[i]  
            echanges += 1  

    print(f"Nombre total de comparaisons: {comparaisons}")
    print(f"Nombre total d'échanges: {echanges}")
    return tab


tab = [6, 2, 7, 1, 9, 5, 10,100,30,23,116]
tri_selection(tab)

print("Le tableau trié est :")
for i in range(len(tab)):
    print("%d" % tab[i])
```

```python

n_values = [1000, 2000, 4000, 8000, 16000, 32000]
times = []


for n in n_values:
    
    tab = list(range(n, 0, -1))
    
    
    start_time = time.time()
    tri_selection(tab)
    end_time = time.time()
    
    
    execution_time = end_time - start_time
    times.append(execution_time)

    print(f"Taille du tableau : {n}, Temps d'exécution : {execution_time:.5f} secondes")
```
